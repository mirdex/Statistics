### 傳統統計學與大數據分析差別
- 傳統統計學(以下簡稱統計學)分為以下兩個部分
    - 敘述統計
        - 資料的收集、整合、呈現、解釋等分析步驟，以數值、表格、圖示來敘述資料概況的方法
    - 推論統計
        - 從樣本資料的分析結果對母體資料的某些特性，做合理的估計與推測
- 大數據分析
    - 21世紀，因為資料量較大，工程界直接以電腦做出各種視覺化並加以分析。
    - 目前說的大數據分析是泛指工程界上早已使用大量數據的統計分析，又稱資料科學```(Data Science)```

### 統計學的專有名詞
- ```母體(Population) : ```擁有共同性質的元素或個體所組成的群體，是分析者所要調查的全體對象集合。通常以大寫字母``` N ```表示
- ```樣本(sample) : ```從母體中抽出部分元素所組成的集合，稱為樣本。通常以小寫字母```n```表示；稱為樣本數。
- ```抽樣 : ```一種程序或方法，說明如何從母體抽出樣本
- ```抽樣誤差 : ```樣本是由母體抽樣得來。但在抽樣過程中因抽樣方法、估計方法不同，而產生的誤差稱為抽樣誤差。
- ```母體參數(Population parameter) : ```描述母體特性的統計測量數
- ```統計量(sample statistic) : ```是母體參數的估計公式，描述樣本特性的測量統計量
- ```實驗 : ```對某種個體做特別處理，以期觀察其反應
- ```普查 : ```針對整個母體進行調查

### 統計學的專有名詞之範例應用

- 例 :　某遊戲公司推出新遊戲，想知道顧客對於新遊戲的反應如何？因此，先推出一段時間，看市場反應。如果接受率在60%以上，那麼將大量推出此新遊戲。經調查後，在200個玩過此遊戲的人中，有125人表示喜愛，那麼這家公司應該推出或暫緩?
- 上例之統計用語
    - 母體 : 所有顧客
    - 樣本 : 200個顧客
    - 參數 : 新產品接受率
    - 要探討的母體參數值 : 60%
    - 樣本統計量德觀察值 : 62.5%($125 \over 200$)
- 上例之統計方法
    - 資料收集、整理 : 市場調查200人中125人接受新遊戲程度
    - 母體參數 : 顧客接受率
    - 抽樣樣本統計量的機率理論 :　市調後，接受率的各種可能狀況
    - 母體參數的估計與檢定 : 市調接受率62.5%來推測未來接受率是否大於60%

### 傳統統計學的內容

- 敘述統計
    - 資料呈現(統計圖)
    - 資料整理(統計表)
    - 資料敘述(中央趨勢的統計量、分散狀態的統計量)
- 隨機變數的機率分配
    - 離散型
        - 二項分配
        - 卜瓦松分配
    - 連續型
        - 常態分配
        - t分配
        - 卡方分配
        - F分配
- 推論統計
    - 估計
        - 點估計
        - 區間估計
    - 檢定
        - 檢定統計量
        - 拒絕域
        - P value
        - 檢定法則

### 資料的蒐集
- 初級資料 : 新蒐集的資料
- 次級資料 : 不需要額外收集，既有的資料；如:資料庫的資料

### 資料的類型
- 性質資料
    - 類別資料(Nominal Data)
        - 使用數字來代表事物的類別，這些數字代表不同的分類，並沒有數字的實質意義。
        - 例 : 性別的分類(M/F 或 1/0 或 男/女)、公司員工的職別
    - 順序資料(Ordinal Data)
        - 使用數字來代表研究中某性質的順序，數字有大小等級之分。
        - 例 : 成績的排名(1/2/3 或 A/B/C 或 甲/乙/丙)
- 數量資料
    - 離散型資料(Discrete Type Data)
        - 以計數或計次的方式所得的資料
        - 例 : 某位候選人的支持人數、某晶圓廠的瑕疵品數
    - 連續型資料
        - 以衡量方式所得的資料，通常受衡量單位影響
        - 例 : 小學生身高
